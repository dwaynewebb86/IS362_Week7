First I will import the necessary modules and additional options for future processing

```python
# Import Modules

 # import pandas and numpy
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sea
import os

 # Set some pandas options for controlling output
pd.set_option('display.notebook_repr_html', False)
pd.set_option('display.max_columns', 10)
pd.set_option('display.max_rows', 10)

 # inline graphics
%matplotlib inline
```

Now I will import the .csv file and print the head for analyzation

```python
mv = "movies.csv"

if not os.path.exists(mv):
    url = "https://raw.githubusercontent.com/dwaynewebb86/IS362_Week7/main/movies.csv"
    ratings = pd.read_csv(url, index_col=0)
    ratings.index.name = "User"
    ratings.to_csv(mv)                       
    print("Downloaded and saved locally")
else:
    ratings = pd.read_csv(mv, index_col=0)
    ratings.index.name = "User"
    print("Loaded from local file")

# Print the top rows of the survey for initial investigation
ratings.head()
```

Now I will show the average ratings for per user and per movie, ignoring empty values where a user has not seen a movie

```python
# average rating for each movie
print("Average rating for each movie:")
print(ratings.mean().round(2))

# average rating for each user
print("\nAverage rating for each user:")
print(ratings.mean(axis=1).round(2))
```

Now I will create a new dataframe with normalized ratings per user
using the formula: x_new = (x - x_min) / (x_max - x_min)

This scales each user's ratings between 0 and 1, where:
- 0 = the lowest rating that user gave
- 1 = the highest rating that user gave

```python
normalized = ratings.apply(
    lambda row: (row - row.min()) / (row.max() - row.min()),
    axis=1
)

print("Normalized ratings:")
print(normalized.round(2))

print("\nAverage normalized rating per movie:")
print(normalized.mean().round(2))

print("\nAverage normalized rating per user:")
print(normalized.mean(axis=1).round(2))
```

# Conclusion
The normalized data provided many advantages including
- Removing user rating bias for those users who rate everything high or low
- Placed all users on a common scale 
- Provided ratings that were easier to compare between participants

There were also some disadvantages to this approach
- Loses the absolute value of ratings making different ratings map to the same normalized value  for a more universal view
- The NaN effected each users score potentially skewing results
- Users with less ratings had less reliable normalized scores


# Extra Credit

Now I will create a new dataframe and standarize the ratins per user instead of normalizing

```python
standardized = ratings.apply(
    lambda row: (row - row.mean()) / row.std(),
    axis=1
)

print("Standardized ratings:")
print(standardized.round(2))

print("\nAverage standardized rating per movie:")
print(standardized.mean().round(2))

print("\nAverage standardized rating per user:")
print(standardized.mean(axis=1).round(2))
```

# Obervation
 - Standardization has no fixed range like normalization making extreme ratings more visible
- The highest and lowest ranked movies Sinners and War Machine remained the same using both methods
